```python
# Load the data for 'New_source1_databank_worldbank_org.csv'
population_data = pd.read_csv('New_source1_databank_worldbank_org.csv')

# 'Czechia' is equivalent to 'Chech Republic'
czechia_population = population_data[population_data['Country Name'] == 'Czechia']

# Define the list of years to iterate through
years = [str(year) + ' [YR' + str(year) + ']' for year in range(2015, 2024)]

# Iterate through each year and fill NaN values in 'Total_Population' for 'Chech Republic' with the corresponding values from Czechia
for year in years:
    df.loc[(df['Country'] == 'Czech Republic') & df['Total_Population'].isna(), 'Total_Population'] = czechia_population[year].values[0]

```
