```python
country_pipeline_1 = (
    idf[(idf['Country']==select)&(idf['Year']<= year_slider)]
    .groupby(['Year','Explained_by_GDP','Explained_by_Health'])[yaxis_happy].max()
    #.to_frame()
    .reset_index()
    .sort_values(by=['Year'])  
    .reset_index(drop=True)
) 
country_pipeline_2 = (
    idf[(idf['Year']<=year_slider)&(idf['Happiness'])]
     .groupby('Year')['Happiness'].median()
)
country_year_pipeline = (country_pipeline_1, country_pipeline_2)
```
