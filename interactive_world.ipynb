```python
df['Kontinent'] = df['Country'].map(country_to_continent)
```

```python
idf = df.interactive()
```

```python
year_slider = pn.widgets.IntSlider(name='Year slider', start=2015, end=2024, step=1, value=2017)
year_slider
```

```python
# Radio buttons for Happiness
yaxis_Happy = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['Happiness', 'Happiness_Rank',],
    button_type='success'
)
```

```python
display(yaxis_Happy)
```

```python
continents = [ 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']#'World',

happy_pipeline = (
    idf[
        (idf.Year >= year_slider) &
        (idf.Kontinent.isin(continents))
    ]
    .groupby(['Kontinent', 'Year'])[yaxis_Happy].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')  
    .reset_index(drop=True)
)
```

```python
happy_pipeline
```

```python
happy_plot = happy_pipeline.hvplot(x = 'Year', by='Kontinent', y=yaxis_Happy,line_width=2, title="Happyness by Continent")
```

```python
happy_plot
```

```python
happy_table = happy_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
```

```python
happy_table
```

```python
happy_vs_GDP_scatterplot_pipeline = (
    idf[
        (idf.Year >= year_slider)
        #& (~ (idf.Kontinent.isin(continents)))
    ]
    .groupby(['Kontinent', 'Year', 'Explained_by_GDP'])['Happiness'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')  
    .reset_index(drop=True)
)
```

```python
#happy_vs_GDP_scatterplot_pipeline
```

```python
happy_vs_GDP_scatterplot = happy_vs_GDP_scatterplot_pipeline.hvplot(x='Explained_by_GDP', 
                                                                y='Happiness', 
                                                                by='Kontinent', 
                                                                size=80, kind="scatter", 
                                                                alpha=0.7,
                                                                legend=True, 
                                                                height=500, 
                                                                width=500)
```

```python
happy_vs_GDP_scatterplot
```

```python
yaxis_happiness_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['Explained_by_Trust', 'Explained_by_Freedom', 'Explained_by_GDP', 'Explained_by_Health', 'Explained_by_Social_support'], 
    button_type='success'
)
```

```python
continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']
```

```python

happyness_source_bar_pipeline = (
    idf[
        (idf.Year == year_slider) &
        (idf.Kontinent.isin(continents_excl_world))
    ]
    .groupby(['Year', 'Kontinent'])[yaxis_happiness_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')  
    .reset_index(drop=True)
)
```

```python
happyness_source_bar_plot =happyness_source_bar_pipeline.hvplot(kind='bar', 
                                                     x='Kontinent', 
                                                     y=yaxis_happiness_source, 
                                                     title='Happyness source by continent')
```

```python
world = pd.read_csv('world_geodata.csv')
```

```python
world=pd.merge(world,df,left_on=['Country'],right_on=['Country'],how='left')
world.isna().sum()
```

```python
## plotting map with folium
urban_area_map = folium.Map( location=[0, 0],zoom_start=2.2)
title_html = '''
             <h3 align="center" style="font-size:20px"><b>Happiness by Country </b></h3>
             <h6 align="center" style="font-size:14px"><b>1985-2016 </b></h6>
             '''
urban_area_map.get_root().html.add_child(folium.Element(title_html))
folium.Choropleth(
    geo_data=world,
    name='choropleth',
    data=world,
    columns=['Kontinent', 'Happiness'],
    key_on='feature.properties.Country',
    fill_color='YlOrRd',
    nan_fill_color='White',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Happiness'
).add_to(urban_area_map)
```

```python
style_function = lambda x: {'fillColor': '#ffffff', 
                            'color':'#000000', 
                            'fillOpacity': 0.1, 
                            'weight': 0.1}
highlight_function = lambda x: {'fillColor': '#000000', 
                                'color':'#000000', 
                                'fillOpacity': 0.50, 
                                'weight': 0.1}
NIL = folium.features.GeoJson(
    world,
    style_function=style_function, 
    control=False,
    highlight_function=highlight_function, 
    tooltip=folium.features.GeoJsonTooltip(
        fields=['Country', 'Happiness'],
        aliases=['Country: ','Happiness: '],
        style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;") 
    )
)
urban_area_map.add_child(NIL)
urban_area_map.keep_in_front(NIL)
folium.LayerControl().add_to(urban_area_map)
urban_area_map
```
