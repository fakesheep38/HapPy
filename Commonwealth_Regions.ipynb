```python
df=pd.read_csv('merged_data_4.csv')
```

```python
# List of ISO codes for the 12 countries
iso_codes_to_update = ['UZ', 'KZ', 'MD', 'TJ', 'RU', 'KG', 'BY', 'AZ', 'TM', 'AM', 'GE', 'UA']

# New region name for these countries
new_region_name = 'Commonwealth of Independent States'  

condition = df['ISO_Code'].isin(iso_codes_to_update)

# Update the "Region" column for the selected rows
df.loc[condition, 'Region'] = new_region_name
```

```python
CIS = df['Region'] == 'Commonwealth of Independent States'
df.loc[CIS,:]
```
