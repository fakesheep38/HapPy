```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Laden Sie die Weltkartendaten
world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))

# Überprüfen Sie die ersten Zeilen der Daten
print(world.head())

# Transformieren Sie die Daten in EPSG:4326
world = world.to_crs(epsg=4326)

# Überprüfen Sie die geografische Ausdehnung
min_x = world['geometry'].bounds['minx'].min()
max_x = world['geometry'].bounds['maxx'].max()
min_y = world['geometry'].bounds['miny'].min()
max_y = world['geometry'].bounds['maxy'].max()

print(f"Min X: {min_x}, Max X: {max_x}")
print(f"Min Y: {min_y}, Max Y: {max_y}")
```

```python
print(world.head())
```

```python
world.plot()
plt.title("Weltkarte in EPSG:4326")
plt.show()
```
