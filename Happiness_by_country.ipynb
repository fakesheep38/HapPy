```python
import pandas as pd
import panel as pn
import numpy as np
import hvplot.pandas
import folium
from panel.interact import interact
pn.extension('tabulator')
pn.extension
```

```python
df = pd.read_csv('merged_data_5.csv', sep=',')
#df.head(1)
```

```python
# cache data to improve dashboard performance
if 'data' not in pn.state.cache.keys():

    df = pd.read_csv('merged_data_5.csv')

    pn.state.cache['data'] = df.copy()

else: 

    df = pn.state.cache['data']
```

```python
idf = df.interactive()
```

```python
year_slider = pn.widgets.IntSlider(name='Year slider', start=2015, end=2023, step=1, value=2015)
year_slider
```

```python
idf.head(2)
```

```python
country_list = list(df['Country'].unique())
```

```python
country_list.sort()
```

```python
# create widget
select = pn.widgets.Select(name='Select Country', options=country_list)
```

```python
country_pipeline_1 = (
    idf[(idf['Country']==select)&(idf['Year']<= year_slider)]
    .groupby(['Year'])[['Happiness_Rank','Happiness','Explained_by_GDP','Explained_by_Health']].max()
    #.to_frame()
    .reset_index()
    .sort_values(by=['Year'])  
    .reset_index(drop=True)
) 
country_pipeline_2 = (
    idf[(idf['Year']<=year_slider)&(idf['Happiness'])]
     .groupby('Year')['Happiness'].median()
)
country_year_pipeline = (country_pipeline_1, country_pipeline_2)
```
