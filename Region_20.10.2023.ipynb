```python
import pycountry
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('merged_data_3.csv')
```

```python
df.head()
```

```python
nan_ISO = df['ISO_Code'].isnull() == True
df.loc[nan_ISO,:]
```

```python
# Create a condition to identify rows with NaN ISO_Code and belonging to Namibia
condition = (df['ISO_Code'].isnull()) & (df['Country'] == 'Namibia')

# Update the ISO_Code for these rows to 'NA'
df.loc[condition, 'ISO_Code'] = 'NA'
```

```python
missing_iso_codes = df[df['ISO_Code'].isna()]
missing_iso_codes
```

```python
iso_to_region_mapping = {
    'CZ': 'Central and Eastern Europe',
    'TW': 'Eastern Asia',
    'HK': 'Eastern Asia',
    'MK': 'Central and Eastern Europe',
    'PS': 'Middle East and Northern Africa',
    'NA': 'Sub-Saharan Africa',
    'TR': 'Middle East and Northern Africa',
    'GM': 'Sub-Saharan Africa',
    'LU': 'Western Europe',
    'GT': 'Latin America and Caribbean',
    'KW': 'Middle East and Northern Africa',
    'BY': 'Central and Eastern Europe',
    'TM': 'Commonwealth of Independent States',
    'CY': 'Western Europe',
    'LY': 'Middle East and Northern Africa',
    'AZ': 'Commonwealth of Independent States',
    'LR': 'Sub-Saharan Africa',
    'CG': 'Sub-Saharan Africa',
    'NE': 'Sub-Saharan Africa',
    'KM': 'Sub-Saharan Africa',
    'SZ': 'Sub-Saharan Africa',
    'MG': 'Sub-Saharan Africa',
    'TD': 'Sub-Saharan Africa',
    'YE': 'Middle East and Northern Africa',
    'MR': 'Sub-Saharan Africa',
    'LS': 'Sub-Saharan Africa',
    'BW': 'Sub-Saharan Africa',
    'RW': 'Sub-Saharan Africa',
    'MV': 'Sub-Saharan Africa',
    'SS': 'Sub-Saharan Africa',
    'TT': 'Latin America and Caribbean',
    'SO': 'Sub-Saharan Africa',
    'BZ': 'Latin America and Caribbean'
}

# Fill the "Region" column only for rows with missing regions
df.loc[df['Region'].isna(), 'Region'] = df['ISO_Code'].map(iso_to_region_mapping)

# Check the DataFrame for any remaining missing values in Region
missing_regions = df[df['Region'].isna()]
```
